```python
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('1/104_sat.jpg')
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
mask = cv2.imread('1/104_mask.png')
```

```python
train_img = np.hstack((img,mask))
```

```python
cv2.imshow('train_image',train_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite('1.jpg',train_img)
```

```python
path = '1/'
```

```python
def pairwise(iterable):
    a = iter(iterable)
    return zip(a,a)
```

```python
d = 1
for _, _ , files in os.walk(path):
    for img1, img2 in pairwise(files):
        mask_img = cv2.imread(path + img1)
        sat_img = cv2.imread(path + img2)
        train_img = np.hstack((sat_img, mask_img))
        filename = "train/%d.jpg"%d
        cv2.imwrite(filename,train_img)
        d+=1
```
